```python
import pdfplumber
import csv

PDF_PATH = "/Users/dana/Desktop/hydrogen_storage/data/3/The datasets used in this work and the corresponding sources.pdf"
CSV_PATH = "table3.csv"

# 1. Extract all table rows from every page of the PDF
rows = []
with pdfplumber.open(PDF_PATH) as pdf:
    for page in pdf.pages:
        for table in page.extract_tables():
            rows.extend(table)

# 2. Drop fully empty rows
cleaned = [row for row in rows if not all(cell in (None, "") for cell in row)]

# 3. Save to CSV
with open(CSV_PATH, "w", newline="") as f:
    csv.writer(f).writerows(cleaned)

print(f"Saved {len(cleaned)} rows to {CSV_PATH}")
```
